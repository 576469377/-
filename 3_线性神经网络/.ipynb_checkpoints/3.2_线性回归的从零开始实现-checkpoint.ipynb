```python
%matplotlib inline
import random
import torch
from d2l import torch as d2l
```

### generate dataset

generate a dataset with 1000 samples, each data has 2 dimension

$$\mathbf{X} \in \mathbb{R}^{1000 \times 2}$$

$$\mathbf{y} = \mathbf{X} \mathbf{w} + b + \epsilon$$

```python
#@save
def synthetic_data(w, b, num_examples):
    """Generate y = Xw + b + noise"""
    X = d2l.normal(0, 1, (num_examples, len(w)))
    y = d2l.matmul(X, w) + b
    y += d2l.normal(0, 0.01, y.shape)
    return X, d2l.reshape(y, (-1, 1))
```

```python
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
```

### data_iter

```python
# use to generate a batch data iter
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    # read the data randomly
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(
            indices[i: min(i + batch_size, num_examples)]
        )
        yield features[batch_indices], labels[batch_indices]
```

```python
batch_size = 10
for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y)
    break
```

### initialization

```python
w = torch.normal(0, 0.01, size=(2, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

### define the model

```python
#@save
def linreg(X, w, b):
    """linear regression model"""
    return torch.matmul(X, w) + b
```

### define the loss function

```python
#@save
def squared_loss(y_hat, y):
    """least mean square loss(LMS)"""
    return (y_hat - y.reshape(y_hat.shape))**2 / 2
```

### define the optimization algorithm

```python
#@save
def sgd(params, lr, batch_size):
    """stachostic gradient descent(mini-batch)"""
    with torch.no_grad():
        for param in params:
            param -= lr * param.grad / batch_size
            param.grad.zero_()
```

### training

```python
lr = 0.03
num_epochs = 3
net = linreg
loss = squared_loss
```

```python
for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        # loss on mini-batch of X and y
        l = loss(net(X, w, b), y)
        l.sum().backward()
        sgd([w, b], lr, batch_size)
    with torch.no_grad():
        train_l = loss(net(features, w, b), labels)
        print(f'epoch {epoch + 1}, loss {float(train_l.mean()):f}')
```
